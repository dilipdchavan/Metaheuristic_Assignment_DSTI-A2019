```python
# Need to Create a matrix of city coordinates
qa_cities_coord = qa_cities[['a','b']].values.tolist()
print(qa_cities_coord)
```

```python
# Initialize function parameters --
population=500
cities_cnt = int(len(qa_cities_coord))
print(cities_cnt)
```

```python
# Need to initialize fitness function object
fitness_coords = mlrose.TravellingSales(coords=qa_cities_coord)
```

```python
#Need to define an Optimization Problem Object with TSPOpt() - For Qatar 194 cities
problem_fit = mlrose.TSPOpt(length=194,fitness_fn=fitness_coords,maximize=False)
```

## Approach1

```python
#To solve Randomised Optimization Algorithm - using the genetic algorithm : 
best_state, best_fitness, fit_curve = mlrose.genetic_alg(problem=problem_fit,
                                                             pop_size=500,
                                                             mutation_prob=0.1,
                                                             max_attempts=100,
                                                             max_iters=1000,
                                                             curve=True)
```

```python
print(best_state)
```

```python
print(best_fitness)
```

```python
print(fit_curve)
```

```python
# TSP - Optimised Route
fig = plt.figure(figsize=(10, 10))
plt.plot([qa_cities_coord[best_state[i % cities_cnt]][0] for i in range(cities_cnt+1)], 
         [qa_cities_coord[best_state[i % cities_cnt]][1] for i in range(cities_cnt+1)], 'ro-', linewidth=1)
```

```python
# TSP - convergence curve
fig = plt.figure(figsize=(10, 8))
plt.plot(-(fit_curve))
```

## Approach2

```python
#To solve Randomised Optimization Algorithm - using the genetic algorithm : 
best_state, best_fitness, fit_curve = mlrose.genetic_alg(problem=problem_fit,
                                                             pop_size=1000,
                                                             mutation_prob=0.2,
                                                             max_attempts=200,
                                                             max_iters=2000,
                                                             curve=True)
```
